```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#import datasets as dataframes
age_df = pd.read_csv("Age-standardized suicide rates.csv")
crude_df = pd.read_csv("Crude suicide rates.csv")
facilities_df = pd.read_csv("Facilities.csv")
hr_df = pd.read_csv("Human Resources.csv")
```

```python
#print head to check for accuracy
age_df.head()
```

```python
#filter age dataframe by gender
male_df = age_df[age_df["Sex"].str.contains("Male")]
female_df = age_df[age_df["Sex"].str.contains("Female")]
```

```python
#print male head to ensure accuracy
male_df.head()
```

```python
#print female head to ensure accuracy
female_df.head()
```

```python
#concatenate male and female to get male and female combined dataframe
male_female_df = pd.concat([male_df, female_df])
```

```python
#print whole dataframe to ensure accuracy
print(male_female_df)
```

```python
#create scatterplot by year of male vs female suicide deaths
sns.relplot(data = male_female_df, x="Year", y="Rate", hue="Sex").set(title="Suicide Rates by Gender")
plt.show()
```

```python
#create scatterplot of male suicide rates by year
plt.scatter("Year", "Rate", data=male_df)
plt.rcParams["figure.figsize"] = [5,5]
plt.title("Male Suicicde Rates")
plt.show()
```

```python
df1 = male_df[male_df["Year"] == 2000]
df2 = male_df[male_df["Year"] == 2010]
df3 = male_df[male_df["Year"] == 2015]
df4 = male_df[male_df["Year"] == 2016]
print("Median value in 2010: ", df1["Rate"].median(),"\nMedian value in 2010: ", df2["Rate"].median(),
     "\nMedian value in 2015: ", df3["Rate"].median(), "\nMedian value in 2016: ", df4["Rate"].median())
```

```python
#create scatterplot of female suicide rates by year
plt.scatter("Year", "Rate", data=female_df, color="red")
plt.rcParams["figure.figsize"] = [5,5]
plt.title("Female Suicide Rates")
plt.show()
```
